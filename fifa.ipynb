```python
import pandas as pd
```

```python
dados = pd.read_csv('fifa.csv')
```

```python
dados
```

```python
dados['total'] = dados['Acceleration'] + dados['Agility'] + dados['Reactions']
dados
```

```python
dados = dados[['Name', 'total']]
dados
```

```python
dados.sort_values('total', ascending = False)
```

```python
dados.to_csv('dadosFifa1.csv', index = False)
```

```python
dados.to_excel('dadosFifa2.xlsx', index = False)
```

```python
dados.to_csv('dadosFifa3.txt', index = False, sep='\t')
```

```python
dados.to_json('dadosFifa4.json', index = False)
```
