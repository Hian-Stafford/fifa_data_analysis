```python
import pandas as pd
```

Lê e mostra os dados na tela

```python
dados = pd.read_csv('fifa.csv')
```

```python
dados
```

Adiciona mais uma coluna com a soma das colunas de Aceleração, Agilidade e Reações dos jogadores

```python
dados['total'] = dados['Acceleration'] + dados['Agility'] + dados['Reactions']
dados
```

Mostra apenas os dados que queremos tirar proveito 

```python
dados = dados[['Name', 'Agility', 'Reactions', 'total']]
dados
```

Traz o valor total de acordo com os 30 jogadores que obtiveram o melhor resultado

```python
dados.sort_values('total', ascending = False).head(30)
```

Exporta os dados de acordo com a extensão

```python
dados.to_csv('dadosFifa1.csv', index = False)
```

```python
dados.to_excel('dadosFifa2.xlsx', index = False)
```

```python
dados.to_csv('dadosFifa3.txt', index = False, sep='\t')
```
